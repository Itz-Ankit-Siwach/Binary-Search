```python
# creating function 
def add_time(st_time,duration,day):
    
    # spliting starting time in two parts 1st time and 2nd am or pm
    h,am_pm=st_time.split(" ") 
    # now spliting 1st part of starting time into hours and minutes
    h,m=h.split(":")
    # type casting into integer to perform addition opertion
    h=int(h)
    m=int(m)
    
    # spliting duration time into hours and minutes
    h1,m1=duration.split(":")
    # now typecasting into integer
    h1=int(h1)
    m1=int(m1)
    
    # h2 is the additional hour that can occur due to addition of minutes leads there sum>60 and increment hour 
    # initially h2 =0
    h2=0
    # calculating final minutes after addition
    final_m=m+m1
    # now checking if minutes sum >60 then how much hour increase and what will be final miute value
    if(final_m>=60):
        h2=final_m//60
        final_m=final_m%60
    
    # calulating sum of all hours from starting time duration and one came after addition of the minutes
    final_h=h+h1+h2
    
    # Calulating if how many days will pass after duration time
    day_count=0
    if(final_h>=24):
        day_count=final_h//24
    
    # Edge cases of calulation of time in 12-hour format like at night after 11:59 pm time will be 00:00 am or in noon after 11:59 am time will be 12:00 pm
    if(final_h%12==0):
        if(am_pm=="PM" or am_pm=="pm"):
            am_pm="AM"
            final_h=0
            day_count=day_count+1
        elif(am_pm=="AM" or am_pm=="am"):
            am_pm="PM"
            final_h=12
    # edge cases of final time after coverting in 12-hour format 
    elif(final_h%12>0):
        if(am_pm=="AM" or am_pm=="am"):
            final_h=final_h%12
            am_pm="PM"
        elif(am_pm=="PM" or am_pm=="pm"):
            final_h=final_h%12
            am_pm="AM"
            day_count=day_count+1
    else:
        final_h=final_h%12
            
    # Coverting day into capital letters to avoid mistake in input from user 
    day=day.upper()
    # creating list of days of week
    day_list=["SUNDAY","MONDAY","TUESDAY","WEDNESDAY","THRUSDAY","FRIDAY","SATURDAY"]
    #calculating the day if dome days had passed
    if(day_count>0):
        # fatching day from list 
        final_day=day_list[(day_list.index(day)+day_count)%7]
        
    # printing final answer
    print(f"The initial time was {st_time} and after duration of {duration} the time is {final_h}:{final_m} {am_pm} and the final day is {final_day} after {day_count} days had passed")
    

    
# Taking inputs from user 
st_time=input("Please enter the starting time in 12-hour format(e.g- 11:44 PM) : ")
duration=input("Enter duration time after you want to check (e.g- 24:34) : " )
day=input("Enter today's day of week (e.g-Sunday) : ")

# Calling function
add_time(st_time,duration,day)
```
